```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('company_sales_data.csv')
df
```

### compare two sales of product with bar chat

```python
facewash=df.facewash
facecream=df.facecream
month=df.month_number
```

```python
month_name=["january","february","march","april", "may","june","july","august","september","october","november","december"]
x_pos=np.arange(len(month))

```

```python

plt.bar(x_pos -0.2,facecream, width=0.4, label='facecream')
plt.bar(x_pos +0.2,facewash, width=0.4,label='facewash')
plt.xticks(x_pos,month)
plt.legend()
plt.show()
```
