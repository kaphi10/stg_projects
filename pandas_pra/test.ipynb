```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('C:/Users/kafayat/Desktop/STUTERN1.3/stg_1_1.3/pandas_pra/company_sales_data.csv')
df
```

```python
month=['jan','feb','march','apr','may','june','july','aug','sept','oct','nov','dec']
x_pos=np.arange(len(month))
facecream=df['facecream']
facewash=df['facewash']
bathingsoap=df['bathingsoap']
```

```python
plt.bar(x_pos -0.2,facecream, width=0.4, label='facecream')
plt.bar(x_pos +0.2,facewash, width=0.4,label='facewash')
plt.xticks(x_pos,month)
plt.xlabel('months')
plt.ylabel('cost of items sold')
plt.title('comparing cost of items sold per months')
plt.legend()
plt.show()
```

```python
mean=df['bathingsoap'].mean()
mean
plt.bar(x_pos,bathingsoap,yerr=mean,capsize=10)
plt.xticks(x_pos,month)
plt.title('sales of bathing soap per month considering the average sales')
plt.xlabel('months')
plt.ylabel('sales of bathing soap')
plt.show()
```
