```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('company_sales_data.csv')
df
```

```python
month=df['month_number']
product1=df['facecream']
product2=df['facewash']
product3=df['toothpaste']
product4=df['bathingsoap']
product5=df['shampoo']
product6=df['moisturizer']

#print(product1,month)
```

```python
plt.plot(month,product1,marker='*',color='green',label='facecream')
plt.plot(month,product2,marker='s',color='red',label='facewash')
plt.plot(month,product3,marker='+',color='black',label='toothpaste')
plt.plot(month,product4,marker='*',color='red',label='bathingsoap')
plt.plot(month,product5,marker='s',color='green',label='shampoo')
plt.plot(month,product6,marker='o',color='black',label='moisturizer')
plt.ylabel("product sales")
plt.xlabel("months")
plt.title("product sales for each specific months")
plt.legend(loc='upper left')
plt.show
```
