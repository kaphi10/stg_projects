```python
#import all liberary
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('anscombe_quartet_data.csv')
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
x1=df.x1
x2=df.x2
x3=df.x3
x4=df.x4
# y axis 
y1=df.y1
y2=df.y2
y3=df.y3
y4=df.y4
```

```python
plt.subplot(2,2,1)
plt.title('first subplot')
plt.scatter(x1,y1,color='green')
#plot line to show the correlation
z1=np.polyfit(x1,y1,1)
p1=np.poly1d(z1)
plt.plot(x1,p1(x1))

plt.subplot(2,2,2)
plt.title('second subplot')
plt.scatter(x2,y2,color='red')
#plot line to show the correlation
z2=np.polyfit(x2,y2,1)
p2=np.poly1d(z2)
plt.plot(x2,p2(x2))


plt.subplot(2,2,3)
plt.title('first subplot')
plt.scatter(x3,y3,color='yellow')
#plot line to show the correlation
z3=np.polyfit(x3,y3,1)
p3=np.poly1d(z3)
plt.plot(x3,p3(x3))


plt.subplot(2,2,4)
plt.title('first subplot')
plt.scatter(x4,y4,color='black')
#plot line to show the correlation
z4=np.polyfit(x4,y4,1)
p4=np.poly1d(z4)
plt.plot(x4,p4(x4))

plt.subplots_adjust(wspace=0.5,hspace=0.5,bottom=0.2,top=0.9)
plt.show()



```

 ### using second approach

```python
#unpack the axis
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2)
ax1.scatter(x1,y1, color='green')
ax1.set_title('first subplot')
ax1.set_xlabel('x1-axis')
ax1.set_ylabel('y1-axis')
#plot line to show the correlation
z=np.polyfit(x1,y1,1)
p=np.poly1d(z)
ax1.plot(x1,p(x1),color='yellow')

ax2.scatter(x2,y2, color='black')
ax2.set_title('second subplot')
ax2.set_xlabel('x2-axis')
ax2.set_ylabel('y2-axis')
#plot line to show the correlation
z=np.polyfit(x2,y2,1)
p=np.poly1d(z)
ax2.plot(x2,p(x2),color='red')

ax3.scatter(x3,y3, color='red')
ax3.set_title('third subplot')
ax3.set_xlabel('x3-axis')
ax3.set_ylabel('y3-axis')
z=np.polyfit(x3,y3,1)
p=np.poly1d(z)
ax3.plot(x3,p(x3),color='black')

ax4.scatter(x4,y4, color='yellow')
ax4.set_title('second subplot')
ax4.set_xlabel('x4-axis')
ax4.set_ylabel('y4-axis')
z=np.polyfit(x4,y4,1)
p=np.poly1d(z)
ax4.plot(x4,p(x4),color='red')
fig.set_figheight(10)
fig.set_figwidth(10)
plt.subplots_adjust(wspace=0.5,hspace=0.5,bottom=0.2,top=0.9)
```

### the third method

```python
fig=plt.figure(figsize=(10,10))
ax5=fig.add_subplot(221,title='subplot1',xlabel='x1-axis',ylabel='y1-axis')
ax5.scatter(x1,y1,color='green')
#plot line to show the correlation
z=np.polyfit(x1,y1,1)
p=np.poly1d(z)
ax5.plot(x1,p(x1),color='yellow')

ax6=fig.add_subplot(222,title='subplot2',xlabel='x2-axis',ylabel='y2-axis')
ax6.scatter(x2,y2,color='red')
#plot line to show the correlation
z=np.polyfit(x2,y2,1)
p=np.poly1d(z)
ax6.plot(x2,p(x2),color='black')

ax7=fig.add_subplot(223,title='subplot3',xlabel='x3-axis',ylabel='y3-axis')
ax7.scatter(x3,y3,color='black')
#plot line to show the correlation
z=np.polyfit(x3,y3,1)
p=np.poly1d(z)
ax7.plot(x3,p(x3),color='red')

ax8=fig.add_subplot(224,title='subplot4',xlabel='x4-axis',ylabel='y4-axis')
ax8.scatter(x4,y4,color='yellow')
#plot line to show the correlation
z=np.polyfit(x4,y4,1)
p=np.poly1d(z)
ax8.plot(x4,p(x4),color='green')

plt.subplots_adjust(wspace=0.5,hspace=0.5,bottom=0.2,top=0.9)
```
